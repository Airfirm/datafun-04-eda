### Section 4. Initial Data Distribution for Numerical Columns

```python
# Inspect histogram by one numerical column
iris_df['sepal_length'].hist()

# Inspect histograms for ALL numerical columns
iris_df.hist()

# Show all plots
matplotlib.pyplot.show()
```

Plotting a histogram of the sepal_length column from the Iris datasets, showing the distribution (frequency) of sepal lengths across all samples.

Plotting histograms for all numerical columns in the dataframe. Each numerical column gets its own histogram, helps to quickly inspect the distribution of each feature.

Displays all the generated plots in the script output.

### Section 4. Initial Data Distribution for Categorical Columns

```python
# Inspect value counts by categorical column
# Column name must be EXACT.
# The value_counts() method is only available for Series objects.
# The value_counts() method returns a pandas Series with the counts of unique values in the column.
iris_df['species'].value_counts()

# Inspect value counts for ALL categorical columns
for col in iris_df.select_dtypes(include=['object', 'category']).columns:
    # Display count plot
    sns.countplot(x=col, data=iris_df)
    matplotlib.pyplot.title(f'Distribution of {col}')
    matplotlib.pyplot.show()

# Show all plots
matplotlib.pyplot.show()
```

Counts the number of occurences of each unique value in the 'species' column.

Returns a pandas 'series' showing how many samples belong to each species (setosa, versicolor, virginica).

### Insights: 
The dataset is perfectly balanced with 50 samples per species. No class imbalance, meaning machine learning models won't be biased toward one species.

### Section 5. Data Transformation and Feature Engineering

```python
# Feature Engineering
# Renaming a column
iris_df.rename(columns={'sepal_length': 'Sepal Length'}, inplace=True)

# Adding a new column
iris_df['Sepal Area'] = iris_df['Sepal Length'] * iris_df['sepal_width']
```

Renaming the column 'sepal_length' to 'Sepal Length' (capitalized with a space) in the Dataframe.

inplace=True applies the change directly to iris_df without needing to reassign it.

### Before: sepal_length
### After: Sepal Length

### Purpose: 
To follow naming conventions (e.g., spaces for display purposes) and improve readability for reports/visualizations.

### Section 6. Initial Visualizations

```python
# Feature Engineering
# Renaming a column
iris_df.rename(columns={'sepal_length': 'Sepal Length'}, inplace=True)

# Adding a new column
iris_df['Sepal Area'] = iris_df['Sepal Length'] * iris_df['sepal_width']

# Create a pairplot of the Iris dataset
# A pairplot is a grid of scatter plots for each pair of numerical columns in the dataset
# The hue parameter is used to color the data points 
# by species (a categorical column)
sns.pairplot(iris_df, hue='species')
plt.suptitle('Pairplot Relationships in Iris Dataset', y=1.02)

# Show all plots
matplotlib.pyplot.show()
```

## Performing data transformation, visual exploration of the Iris dataset.
##### Generate a pairwise Scatter plots (diagonal) and histograms (diagonal) for all numerical columns (Sepal Length, sepal_width, petal_length, petal_width, Sepal Area).
##### Colors points by species
##### Diagonal shows histograms of each feature
##### Insights: Strong correlation between petal length/width, and clear species separation in petal.
##    Species Separation:
###   Petal Features:
##### Separation between species: setosa has smaller petals than others.
##### Strong positive correlation (points from a diagonal line).
###   Sepal Features:
##### setosa has smaller Sepal Length but larger sepal_width compared to others.
##### Sepal Area helps differentiate setosa (compact sepals) from virginica (larger sepals).
##    Feature Correlation:
##### Petal dimensions are highly predictive of species (minimal overlap in scatterplots).
####  Outliers: Check for points that deviate from species clusters (potential data errors or rare variants). 


```python
# A scatter plot is a plot of two numerical variables.
scatter_plt: Axes = sns.scatterplot(
    data=iris_df, x="Sepal Length", y="Sepal Area", hue="species"
)

# Set axis labels using the Matplotlib Axes methods set_xlabel() and set_ylabel()
scatter_plt.set_xlabel("Sepal Length (mm)")
scatter_plt.set_ylabel("Sepal Area (mm squared)")   

# Set the title using the Matplotlib Axes set_title() method
scatter_plt.set_title("Chart 1. Iris Sepal Length vs. Sepal Area (by Species)")

matplotlib.pyplot.show()
```

###   Created a scatter plot to visualize the relationship between Sepal Length and Sepal Area in the Iris dataset, with points colored by species.
###   Species Specific Patterns:
##### setosa (blue points): short sepals but large area - compact, wide sepals (high width compensates for short length).
##### virginica (green points): long sepals but variable area - some have large areas (long + wide), others are narrower.
##### versicolor (orange points): intermediate traits - falls between setosa and virginica.
###   Correlation Trends:
##### Positive Correlation: Longer sepals generally have larger areas, but the relationship is stronger in virginica than setosa.
##### Outliers: Points deviating from their species cluster may indicate measurement errors or natural variation.
###   Biological Implications:
##### Sepal shape (length vs width trade-off) may relate to species adaptation (e.g., setosa's compact sepals might aid in drought resistance).

```python
sns.boxplot(data=iris_df, x='species', y='petal_length')
plt.title('Petal Length Distribution by Species')
plt.show()
```

Displays the median (line), quartile (box), and outliers (dots) of petal length.
Insights: virginica has the longest petals, wjile setosa has the shortest, with minimal overlap.

```python
# Create a scatter plot of Sepal Length vs. Sepal Width, colored by species

sns.scatterplot(data=iris_df, x='Sepal Length', y='sepal_width', hue='species')
plt.title('Sepal Length vs. Sepal Width by Species')
plt.show()
```

Displays the relationship between sepal length and width.
Color-coded by species (setosa, versicolor, virginica)
Insights: setosa tends to have shorter sepals but greater width compared to others.
